```python
#Assignment2
##เรียกใช้ library pandas
import pandas as pd
```

```python
##นำเข้าข้อมูลไฟป่าจากGithubและอ่านข้อมูล
fromGithubb = pd.read_excel('https://github.com/bank750/geodatasciencee/raw/master/Fire_Data.15240.1598195125.4438.xlsx')

#ใส่เงื่อนไขเลือกเเค่5วัน
fromGithub = fromGithubb[fromGithubb.ACQ_DATE < '2019-12-06'].copy()
```

```python
##จัดเเต่งข้อมูล/แปลงข้อมูล
fromGithub['ACQ_DATE'] = fromGithub.ACQ_DATE.astype(str)
fromGithub.ACQ_TIME= pd.to_datetime(fromGithub.ACQ_TIME,format='%H%M')
fromGithub['ACQ_TIME'] = fromGithub.ACQ_TIME.astype(str)
fromGithub[['q','ACQ_TIME']] = fromGithub['ACQ_TIME'].str.split('1-01',expand=True)
combined = fromGithub.ACQ_DATE.str.cat(fromGithub.ACQ_TIME,sep = ' ')
fromGithub['combined'] =  pd.to_datetime(combined)
fromGithub['combined'] =  pd.to_numeric(fromGithub['combined'])
```

```python
##plotข้อมูล
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
threedee = plt.figure().gca(projection='3d')
threedee.scatter( fromGithub['LONGITUDE'], fromGithub['LATITUDE'], fromGithub['combined'] )
threedee.set_xlabel('LONGITUDE')
threedee.set_ylabel('LATITUDE ')
threedee.set_zlabel('combined')
plt.show()
```
